# Matplotlib

`matplotlib` é um pacote que oferece ferramentas de plotagem similares ao MatLab.

Aprendemos a importar pacotes no nosso código Python através de 'apelidos'.

No caso do `matplotlib`, incialmente, vamos utilizar apenas o módulo `pyplot`, por isso, vamos importar apenas ele como `plt` (que é um 'apelido' convenção)

```Python
import matplotlib.pyplot as plt #IMPORTAR matplotlib.pyplot COMO plt
```

Vamos também importar o `numpy` como `np`

```Python
import numpy as np 
```

Para plotar uma certa $y(x)$ contínua em um domínio, independente dos limites do intervalo, precisaríamos de um vetor com infinitos elementos, já que existem infinitos números em qualquer intervalo real não degenerado (intervalo que não comece e acabe no mesmo número).

Por isso, vamos discretizar o intervalo contínuo (e a função), em um número finito de pontos. 

Vamos criar um vetor `x` com `1000` elementos usando `np.linspace`:

```Python
x=np.linspace(-10,10,1000) #intervalo fechado entre -10 e 10 com 1000 elementos espaçados uniformemente
```

````python
import matplotlib.pyplot as plt #IMPORTAR matplotlib.pyplot COMO plt
import numpy as np

x=np.linspace(-10,10,1000) #intervalo fechado entre 1 e 2 com 1000 elementos espaçados uniformemente

plt.plot(x, x**2)
````

Vamos tentar plotar um seno:

## Estilização do gráfico


Um jeito rápido de estilizar os gráfico é usando o parâmetro de formatação

`'[estilo do marcador][estilo da linha][cor]'`

Onde os colchetes e tudo dentro deles são substituídos por códigos que estarão listados mais abaixo. Vamos tentar plotar o mesmo gráfico, mas agora com pontos vermelhos ao invés de uma linha azul.

Podemos fazer isso apenas acrescentando um último argumento na função plot

`plt.plot(x, np.sin(y))`$\rightarrow$`plt.plot(x, np.sin(y), 'o--r')`

### Cores Base

Essas são as cores que podem ser utilizadas em um único parâmetro do tipo: `'[estilo do marcador][estilo da linha][cor]'`

![files/colors1.png](files/colors1.webp)

Vamos tentar plotar o mesmo seno, mas apenas com marcadores em formato de x que tem o código `'x'` e na cor magenta que tem o código `'m'`.

````python
import matplotlib.pyplot as plt #IMPORTAR matplotlib.pyplot COMO plt
import numpy as np
N=100
x=np.linspace(-10,10,N) #intervalo fechado entre 1 e 2 com 1000 elementos espaçados uniformemente
plt.plot(x, np.sin(x), 'xm')
````

# Cores CSS

Existem outros esquemas de cores além do base que precisam de um parâmetro de palavra chave específico na função chamado `color`.

![files/colorscss.png](files/colorscss.png)

Por exemplo, podemos repetir o gráfico anterior, porém com a cor css `'coral'`

````python
import matplotlib.pyplot as plt #IMPORTAR matplotlib.pyplot COMO plt
import numpy as np
N=100
x=np.linspace(-10,10,N) #intervalo fechado entre 1 e 2 com 1000 elementos espaçados uniformemente
plt.plot(x, np.sin(x), 'o--', color='coral')
````

Cores base também podem ser especificadas com o parâmetro `color`.
